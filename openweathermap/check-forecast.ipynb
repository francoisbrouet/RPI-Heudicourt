```python
import requests
import datetime
```

```python
lat = 48.931663
lon = 5.697698
api_key = '8db111354c0316b6d76b8ed55b84d8cc'
```

```python
url = 'https://api.openweathermap.org/data/3.0/onecall'
params = {'lat': lat, 'lon': lon, 'exclude': 'daily,alerts', 'appid': api_key, 'units': 'metric'}
```

```python
response = requests.get(url, params=params, timeout=5)
data = response.json()
```

```python
data['hourly']
```

```python
for h in data['hourly']:
    dt = datetime.datetime.fromtimestamp(h['dt'])
    temp = h['temp']
    humidity = h['humidity']
    clouds = h['clouds']
    uv = h['uvi']
    print (dt.strftime('%Y-%m-%d %H:%M'), f'{temp:.1f} °C, {humidity}% humidity, {clouds}% clouds, {uv} uv')
```
